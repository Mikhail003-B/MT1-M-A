## Предобработка целевой характеристики

```python
import re
import numpy as np
```

```python
df['Kd (nM)'] = df['Kd (nM)'].astype(float)
```

```python
# Используем регулярное выражение для поиска нечисловых данных
non_numeric_data = df['Kd (nM)'][~df['Kd (nM)'].astype(str).str.match(r'^-?\d*\.?\d+$')]
unique_non_numeric_data = set(non_numeric_data)
# Вывод нечисловых данных
print("Нечисловые данные в колонке 'Kd (nM)':")
print(unique_non_numeric_data)
```

```python
def replace_non_numeric_with_none(value):
    try:
        float(value)
        return value  # Возвращаем значение, если оно числовое
    except ValueError:
        return None  # Возвращаем None, если значение не числовое

df['Kd (nM)'] = df['Kd (nM)'].apply(replace_non_numeric_with_none)
df['Kd (nM)'] = df['Kd (nM)'].astype(float)
df.info()
```

```python
def extract_number(affinity_str):
    # Регулярное выражение для поиска чисел и их диапазонов
    number_pattern = re.compile(r'Kd(\d+(\.\d+)?)(?:-(\d+(\.\d+)?))?\b')

    # Ищем все совпадения чисел в строке
    matches = number_pattern.finditer(affinity_str)

    for match in matches:
        # Извлекаем найденные числа
        start_value = float(match.group(1))
        end_value = float(match.group(3)) if match.group(3) else None

        # Если это диапазон, берем среднее значение
        if end_value is not None:
            value = (start_value + end_value) / 2
        else:
            value = start_value

        # Проверяем, если после числа стоит "µM"
        unit_match = re.search(r'µM', affinity_str[match.end():])
        if unit_match:
            value *= 1000
            return value

        return value

    return np.nan


# Применение функции к колонке 'Affinity'
df['Affinity_cleaned'] = df['Affinity'].apply(extract_number)
df['Affinity_cleaned']
```

```python
df.info()
```

```python
# Заполнение пропусков в колонке 'Kd (nM)' значениями из 'Affinity_cleaned'
df['Kd (nM)'] = df['Kd (nM)'].fillna(df['Affinity_cleaned'])

# Удаление временной колонки 'Affinity_cleaned'
df.drop(columns=['Affinity_cleaned'], inplace=True)

# Итоговый DataFrame
print(df)
```

```python
import requests
```

```python
# Функция для получения UniProt ID по названию белка
def get_uniprot_id(protein_name):
    search_url = "https://rest.uniprot.org/uniprotkb/search"
    params = {
        'query': protein_name,
        'format': 'json',
        'size': 1  # Ограничиваем количество результатов до 1
    }
    response = requests.get(search_url, params=params)
    if response.status_code == 200:
        results = response.json()
        if results['results']:
            return results['results'][0]['primaryAccession']
    return None

# Функция для получения аминокислотной последовательности по UniProt ID
def get_uniprot_sequence(uniprot_id):
    sequence_url = f"https://rest.uniprot.org/uniprotkb/{uniprot_id}.fasta"
    response = requests.get(sequence_url)
    if response.status_code == 200:
        return response.text
    return None

# Функция для получения аминокислотной последовательности по названию белка
def fetch_protein_sequence(protein_name):
    uniprot_id = get_uniprot_id(protein_name)
    if uniprot_id:
        return get_uniprot_sequence(uniprot_id)
    return None
```

```python
df['target_sequence'] = df['Target '].apply(fetch_protein_sequence)
df.info()
```

```python
df['target_sequence'][1]
```

```python
def filter_sequence(sequence):
    if sequence:
        parts = sequence.split('\n', 1)  # Разделяем строку по первому '\n'
        if len(parts) > 1:
            filtered_sequence = re.sub(r'\n', '', parts[1])  # Убираем все '\n' из второй части
            return filtered_sequence
    return None

# Применение функции к колонке 'target_sequence'
df['filtered_sequence'] = df['target_sequence'].apply(filter_sequence)
```

```python
Entrez.email = "evgenynam22@gmail.com"

# Функция для получения GI номера по названию белка
def get_ncbi_gi(protein_name):
    search_url = "https://eutils.ncbi.nlm.nih.gov/entrez/eutils/esearch.fcgi"
    params = {
        'db': 'protein',
        'term': protein_name,
        'retmode': 'json',
        'retmax': 1  # Ограничиваем количество результатов до 1
    }
    response = requests.get(search_url, params=params)
    if response.status_code == 200:
        results = response.json()
        if results['esearchresult']['idlist']:
            return results['esearchresult']['idlist'][0]
    return None

# Функция для получения аминокислотной последовательности по GI номеру
def get_ncbi_sequence(gi_number):
    fetch_url = "https://eutils.ncbi.nlm.nih.gov/entrez/eutils/efetch.fcgi"
    params = {
        'db': 'protein',
        'id': gi_number,
        'rettype': 'fasta',
        'retmode': 'text'
    }
    response = requests.get(fetch_url, params=params)
    if response.status_code == 200:
        return response.text
    return None

# Функция для получения аминокислотной последовательности по названию белка
def fetch_protein_sequence_ncbi(protein_name):
    gi_number = get_ncbi_gi(protein_name)
    if gi_number:
        return get_ncbi_sequence(gi_number)
    return None
```

```python
def extract_ph_value(buffer_conditions):
    match = re.search(r'pH\s*([\d\.]+)', buffer_conditions)
    if match:
        return float(match.group(1))
    return None

# Применение функции к колонке 'Binding Buffer/Conditions'
df['pH'] = df.apply(
    lambda row: extract_ph_value(
        row['Binding Buffer/Conditions']) if row['pH'] is None else row['pH'], axis=1
    )
df.info()
```

```python
# Заполнение пропущенных значений средним значением колонки pH_value
mean_ph = df['pH'].mean()
df['pH'].fillna(mean_ph, inplace=True)
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder

# Создание экземпляра LabelEncoder
le_nucleic_acid = LabelEncoder()
le_buffer = LabelEncoder()

# Кодирование колонок
df['Type of Nucleic Acid'] = le_nucleic_acid.fit_transform(df['Type of Nucleic Acid'])
df['Type of the buffer'] = le_buffer.fit_transform(df['Type of the buffer'])
df
```
